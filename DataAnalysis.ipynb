# Age

```python
mean_age = df['age'].mean()
mean_age
```

```python
mean_age = df['age'].mean()
median_age = df['age'].median()
mode_age = df['age'].mode()[0]

print(f"Mean age: {mean_age}")
print(f"Median age: {median_age}")
print(f"Mode age: {mode_age}")
```

```python
min(df['age'])
```

```python
import numpy as np
age_data = df['age']

variance = np.var(age_data, ddof=0)# variance, ddof set to 0 to take from entire population
print("Variance:", variance)
std_dev = np.std(age_data, ddof=0)# Standard deviation
print("Standard Deviation:", std_dev)
percentiles = np.percentile(age_data, [25, 50, 75]) # Percentiles (25th, 50th (Median), 75th)
print("25th Percentile:", percentiles[0])
print("50th Percentile (Median):", percentiles[1])
print("75th Percentile:", percentiles[2])
```

```python
df = pd.read_csv('32130_AT2_25526979.csv')
df
```

```python
#numBins = 22
n = len(df)
numBins = int(1 + np.log2(n)) #sturges rule
numBins
```

```python
min = df['height'].min()
min
```

```python
max = df['height'].max()
max
```

```python
binSize = (max-min)/numBins
print("The size of a bin is ",binSize)
```

```python
binEdges = np.linspace(min, max, numBins + 1)  # linspace is used to create fixed width intervals
binEdges
```

```python
#Equi-Width Binning
df["Height_EWB"] = pd.cut(df["height"], bins=binEdges, labels=False, include_lowest=True) #pd.cut assigns automatically according to the bins
df["Height_EWB"]
```

```python
import matplotlib.pyplot as plt

plt.hist(df["Height_EWB"], bins=numBins, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title(f"Histogram of 'Height_EWB' with {numBins} bins")
plt.show()
```

```python
#Equi depth binning
df['quartilebin'] = pd.qcut(df['height'], q=4)
print(df['quartilebin'].value_counts())
df
```

# Discretising age

```python
df = pd.read_csv('32130_AT2_25526979.csv')
df['age'] = df['age']/365
df['age']
```

```python
bins = [0, 21, 34, 44, 64, float('inf')]
labels = ["Young", "Early Adulthood", "Early Middle Age", "Late Middle Age", "Late Adulthood"]
df["ageCategory"] = pd.cut(df["age"], bins=bins, labels=labels, right=True) #using binning
```

```python
ageCategory_freq = df["ageCategory"].value_counts()
print(ageCategory_freq)
# Save to an Excel file
df.to_excel("ageCategory.xlsx", index=False, engine="openpyxl")
```
